```python
with open("/home/niek/HSA_data/header", 'r') as hf:
    header = hf.readline().split(',')
header[-1] = header[-1][:-1]#remove newline
print(header)
```

```python
filename = "/home/niek/HSA_data/data_experiment_1_2.h5"
store = pd.HDFStore(filename)
print("before: ", store)
i = 0
for key in header:
    col = pd.read_csv("/home/niek/HSA_data/data_experiment_1_2.csv", usecols=[key])
    if col[key].dtype == 'object':
        col[key] = col[key].astype(str)
    store[key] = col
    i += 1
print("after: ", store)
```

```python
#remove spaces in keys
for key in header:
    newkey = key
    if key[0] == ' ':
        newkey = newkey[1:]
    if key[-1] == ' ':
        newkey = newkey[:-1]
    newkey = newkey.replace(' ', '_')
    store.get_node(key)._f_rename(newkey)
print(store)
```
