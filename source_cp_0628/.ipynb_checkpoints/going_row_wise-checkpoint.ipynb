```python
#count occurences of each 'Run'
counter = {}
for chunk in chunks:
    for i,row in chunk.iterrows():
        if not row['Run'] in counter.keys():
            counter[row['Run']] = 1
        else:
            counter[row['Run']] += 1
print(counter)
```

```python
#count occurences of each 'Scan' in one 'Run'
counter = {}
run = 'Belsom_Rappsilber_PurHSA_01'
for chunk in chunks:
    for i,row in chunk.iterrows():
        if not row['Run'] == run:
            continue
        if not row['Scan'] in counter.keys():
            counter[row['Scan']] = 1
        else:
            counter[row['Scan']] += 1
print(counter)
```

```python
def getFoundRatio(dic):
    foundCorrect = 0
    count = 0
    if len(dic) == 0:
        return 0.
    for key in dic:
        for match in dic[key]:
            count += 1
            if match[1] < 8.:
                foundCorrect += 1
    return float(foundCorrect) / count
```

```python
#count number of different experiments
runs = set()
for chunk in chunks:
    for i,row in chunk.iterrows():
        runs.add(row['Run'])
nofruns = len(runs)
```

```python
chunks = pd.read_csv("/home/niek/HSA_data/data_experiment_1_2_nodecoys.csv", usecols=columns, chunksize=1e5)
```

```python
#top match scores
contactFoundRatio = {}#for each experiment the percentage of found matches that correspond to a true contact
for ex in runs:
    scans = {}
    chunks = pd.read_csv("/home/niek/HSA_data/data_experiment_1_2_nodecoys.csv", usecols=columns, chunksize=1e5)
    for chunk in chunks:
        for i,row in chunk.iterrows():
            if row['Run'] != ex:
                continue
            if not row['Scan'] in scans.keys():
                scans[row['Scan']] = []
            realdist = 1.
            scans[row['Scan']] += [(row['match score'], realdist)]
    for scan in scans.keys():
        scans[scan] = sorted(scans[scan], key=lambda match: match[0], reverse=True)#descending by score
    contactFoundRatio[ex] = getFoundRatio(scans)
```
